```python
import pandas as pd
```

```python
# Importando arquivo
aluguel_casa = pd.read_excel('houses_rent.xlsx')
```

```python
# Visualizando as 5 primeiras linhas
aluguel_casa.head()
```

```python
# Visualizando as últimas 5 linhas
aluguel_casa.tail()
```

```python
# Renomeando o nome das colunas
aluguel_casa = aluguel_casa.rename(columns={'parking spaces':'parking_spaces', 'rent amount':'rent_amount', 'property tax':'property_tax', 'fire insurance':'fire_insurance'})
```

```python
aluguel_casa.head()
```

```python
# Visualizando os tipos de dados
aluguel_casa.dtypes
```

```python
# Alterando o tipo de dados
aluguel_casa = aluguel_casa.astype({'rent_amount':'float', 'property_tax':'float', 'fire_insurance':'float', 'total': 'float'})
aluguel_casa.dtypes
```

```python
aluguel_casa.head()
```

```python
# Informação dos dados
aluguel_casa.info()
```

```python
# Selecionando as colunas desejadas e descrevendo os dados
aluguel_casa[['area', 'rooms', 'bathroom', 'parking_spaces', 'rent_amount', 'property_tax', 'fire_insurance', 'total']].describe()
```

```python
# Filtro por valor abaixo de 4000
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000)]
aluguel_casa.head()
```

```python
# Valor máximo do filtro
aluguel_casa.describe()
```

```python
# Filtrando o aluguel com valor menor que 4000 e que aceita pet
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept')]
aluguel_casa.head()
```

```python
# Filtrando o aluguel com valor menor que 4000, que aceita pet e com 2 e 3 quartos
aluguel_casa = aluguel_casa[(aluguel_casa['total']<=4000) & (aluguel_casa['animal']=='accept') & ((aluguel_casa['rooms']==2) | (aluguel_casa['rooms']==3))]
aluguel_casa.head()
```

```python
# Filtro SQL valor menor que 4000 e aceita animal
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
# Filtrando com SQL o aluguel com valor menor que 4000, que aceita pet e com 2 e 3 quartos
aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3")
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
# Agregação - agrupar por cidade
agg_city_price = aluguel_casa.groupby(['city'])['total'].agg('mean').reset_index()
agg_city_price.head()
```

```python
# Agregação - agrupar por banheiro e quarto
agg_bath_room_price = aluguel_casa.groupby(['bathroom', 'rooms'])['total'].agg('mean').reset_index()
agg_bath_room_price.head(10)
```

```python
# Importando o seaborn para plotar
import seaborn as sns
```

```python
# Plotando média de preços dos quartos
ax = sns.barplot(data=agg_bath_room_price, x='rooms', y='total', color='blue');
ax.set(xlabel='Qtd de quartos', ylabel='Média de preço', title='Preço por quartos');
```

```python
# Plotando média de preços dos banheiros
ax = sns.barplot(data=agg_bath_room_price, x='bathroom', y='total', color='orange');
ax.set(xlabel='Qtd de banheiro', ylabel='Média de preço', title='Preço por banheiro');
sns.set_theme(style='darkgrid');
```
